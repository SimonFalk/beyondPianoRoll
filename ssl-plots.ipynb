```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
#60 34.62 ± 0.19 33.70 ± 6.26
#300 60.04 ± 3.07 65.27 ± 3.41
#600 66.41 ± 4.93 75.17 ± 1.52
#1500 82.91 ± 1.92 85.93 ± 0.98
#4892 93.37 ± 0.71 93.41 ± 0.32
```

```python
x = np.array([60,300,600,1500,4892])
```

```python
y1 = np.array([34.62, 60.04, 66.41, 82.91, 93.37])
s1 = np.array([0.19, 3.07, 4.93, 1.92, 0.71])
```

```python
y2 = np.array([33.70, 65.27, 75.17, 85.93, 93.41])
s2 = np.array([6.26, 3.41, 1.52, 0.98, 0.32])
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_xticks(x)
ax.set_xticklabels([str(lab) for lab in x])
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.errorbar(x,y1, s1)
ax.errorbar(x,y2, s2)
ax.set_xlabel("Number of labeled training samples")
ax.set_ylabel("Classification acc. on test set")
ax.legend(["Supervised", "Mean teacher"])
```
