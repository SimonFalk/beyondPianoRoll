```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
# Signal (violin)
wav_path = 'datasets/initslurtest_vn/initslurtest_vn_wav/slurtest04.wav'
signal, sample_rate = librosa.load(wav_path, sr=44100)
plt.figure(figsize=(25,5))
plt.plot(signal)
plt.show()
```

```python
# Noise
noise_sigma = 2
noise = np.random.normal(scale=noise_sigma, size=signal.shape)
plt.figure(figsize=(25,5))
plt.plot(noise)
plt.show()
```

```python
signal_energy = librosa.feature.rms(signal)
noise_energy = librosa.feature.rms(noise)
```

```python
# Calculate energy

signal_energy = librosa.feature.rms(signal).reshape((-1))
noise_energy = librosa.feature.rms(noise).reshape((-1))

signal_energy.shape
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(signal_energy)
axs[1].plot(noise_energy)
```
